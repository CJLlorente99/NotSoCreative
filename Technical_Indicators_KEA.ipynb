```python
import pandas as pd
import datetime as dt
import talib as ta
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as web
```

```python
start_date=dt.datetime(2018, 7, 1)
end_date=dt.datetime(2018, 9, 26)
data = web.DataReader("^GSPC", "yahoo",start_date,end_date)
```

```python
data
```

```python
data["SMA_2"]=ta.SMA(data["Open"],2)
data["SMA_3"]=ta.SMA(data["Open"],3)
data["SMA_50"]=ta.SMA(data["Open"],50)
data["SMA_100"]=ta.SMA(data["Open"],100)
data["EMA_2"]=ta.EMA(data["Open"],2)
data["EMA_3"]=ta.EMA(data["Open"],3)
data["EMA_50"]=ta.EMA(data["Open"],50)
data["EMA_100"]=ta.EMA(data["Open"],100)
data["RSI"]=ta.RSI(data["Open"],14)
```

```python
data
```

```python
macd, macd_signal, macd_hist =ta.MACD(data["Open"])
```

```python
data["macd"]=macd
data["macd_signal"]=macd_signal
data["macd_hist"]=macd_hist
```

```python
data
```

```python
plt.figure(figsize=(16,8))
plt.title("name plt")
plt.xlabel("Date",fontsize=18)
plt.ylabel("y-label", fontsize=18)
plt.plot(data["SMA_2"]) 
plt.plot(data["SMA_3"])
# plt.plot(data["Open"])
plt.legend(["SMA_2","SMA_3","Open-price"], loc="lower right")
plt.show()
```

```python
data["Kaufsignal_SMA_short"]=""
data["Kaufsignal_SMA_long"]=""
data["Kaufsignal_EMA_short"]=""
data["Kaufsignal_EMA_short"]=""
data["Kaufsignal_EMA_long"]=""
data["Kaufsignal_MACD"]=""
```

```python
data
```

```python
#Crossover fÃ¼r SMA - Short

for i in range(len(data)):
    if ((data.SMA_2[i-1]<data.SMA_3[i-1]) & (data.SMA_2[i]>data.SMA_3[i])):
        data.at[data.index[i],"Kaufsignal_SMA_short"]=1
    elif ((data.SMA_2[i-1]>data.SMA_3[i-1]) & (data.SMA_2[i]<data.SMA_3[i])):
        data.at[data.index[i],"Kaufsignal_SMA_short"]=-1
    else:
        data.at[data.index[i],"Kaufsignal_SMA_short"]=0
     
```

```python
#Crossover fÃ¼r SMA - long

for i in range(len(data)):
    if ((data.SMA_50[i-1]<data.SMA_100[i-1]) & (data.SMA_50[i]>data.SMA_100[i])):
        data.at[data.index[i],"Kaufsignal_SMA_long"]=1
    elif ((data.SMA_50[i-1]>data.SMA_100[i-1]) & (data.SMA_50[i]<data.SMA_100[i])):
        data.at[data.index[i],"Kaufsignal_SMA_long"]=-1
    else:
        data.at[data.index[i],"Kaufsignal_SMA_long"]=0
    
     
```

```python
#Crossover fÃ¼r EMA - Long

for i in range(len(data)):
    if ((data.EMA_50[i-1]<data.EMA_100[i-1]) & (data.EMA_50[i]>data.EMA_100[i])):
        data.at[data.index[i],"Kaufsignal_EMA_long"]=1
    elif ((data.EMA_50[i-1]>data.EMA_100[i-1]) & (data.EMA_50[i]<data.EMA_100[i])):
        data.at[data.index[i],"Kaufsignal_EMA_long"]=-1
    else:
        data.at[data.index[i],"Kaufsignal_EMA_long"]=0
     
```

```python
#Crossover fÃ¼r MACD

for i in range(len(data)):
    if ((data.macd[i-1]<data.macd_signal[i-1]) & (data.macd[i]>data.macd_signal[i])):
        data.at[data.index[i],"Kaufsignal_MACD"]= 1
    elif ((data.macd[i-1]>data.macd_signal[i-1]) & (data.macd[i]<data.macd_signal[i])):
        data.at[data.index[i],"Kaufsignal_MACD"]=-1
    else:
        data.at[data.index[i],"Kaufsignal_MACD"]=0     
```

```python
sma_s_g_ist=[0]*data.shape[0]
```

```python
# Defining buy,hold & sell-function
# Gain for SMA-short
# Unterscheide in theoretischen gain und realisierter gain
sma_s_g_ist=[0]*data.shape[0]
sma_s_g_realisiert=[]

for i in range(len(data)):
    
    if data.Kaufsingal_SMA_short[i]==-1:
        sma_s_g_ist[i]=sma_s_g_ist[i-1]*(data.Open[i]/data.Open[i-1])
        
    elif data.Kaufsingal_SMA_short[i]==1:
        sma_s_g_ist[i]=sma_s_g_ist[i-1]*(data.Open[i]/data.Open[i-1])
    
    #FÃ¼r den Fall dass an der ersten Stelle verkauft werden soll (was nicht mÃ¶glich ist)
    elif (i==0) & data.Kaufsingal_SMA_short[i]==-1:
        sma_s_g_ist[i]=sma_s_g_ist[i-1]
        
    else:
        sma_s_g_ist.append(i)=sma_s_g_ist[i-1]
    
```

```python
sma_s_g_list=[0]*data.shape[0]
save_price=[0]*data.shape[0]

for i in range(len(data)):
    if (data.Kaufsignal_SMA_short[i]==1) & (sma_s_g_ist[i-1]==0)
        save_price[i]=data.Open[i]
    elif (data.Kaufsignal_SMA_short[i]==1) & (save_price[i-1]>0)
        sma_s_g_ist[i]=(data.Open[i]/save_price[i-1])*sma_s_g_list
        
        (data.Open[i]/sma_s_g-ist[i-1])
    
```
